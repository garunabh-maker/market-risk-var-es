```python
import pandas as pd
import numpy as np

import os
print(os.getcwd())

df=pd.read_csv("Hist.csv")

df=df.dropna()
df=df.drop_duplicates(subset='Date')
print(df.head())
print(df.describe())

df=df.sort_index(ignore_index=True)

df[df["Open"].isna()]["Open"].head(10)

df["Open"] = (
    df["Open"]
    .str.replace("$", "", regex=False).astype(float))
df["Close"] = (
    df["Close"]
    .str.replace("$", "", regex=False).astype(float))
df["return"]=(df["Open"]/df["Close"])-1
print(df.head())

return1 = df["return"]
```
